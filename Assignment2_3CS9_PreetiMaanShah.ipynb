## Part 2: Data Preprocessing and Transformation

```python
df.isnull().sum()
```

__Normalization__

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df.describe()
```

```python
mms = MinMaxScaler()
df['YearlyIncome'] = mms.fit_transform(df['YearlyIncome'].values.reshape(-1,1))
```

```python
df['AvgMonthSpend'] = mms.fit_transform(df['AvgMonthSpend'].values.reshape(-1,1))
```

```python
df.describe()
```

__discretization__

```python
from sklearn.preprocessing import KBinsDiscretizer
kb = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
df['YearlyIncome'] = kb.fit_transform(df['YearlyIncome'].values.reshape(-1,1))
```

```python
df['AvgMonthSpend'] = kb.fit_transform(df['AvgMonthSpend'].values.reshape(-1,1))
```

```python
df.describe()
```

__Standardization__

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df[['YearlyIncome','AvgMonthSpend']] = sc.fit_transform(df[['YearlyIncome','AvgMonthSpend']])
```

```python
df.describe()
```

__Binarization/One Hot Encoding__

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse = False)
ohe_features = ['Gender', 'MaritalStatus', 'Education','Occupation','CountryRegionName']
encoded_features = ohe.fit_transform(df[ohe_features])
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out())
df_final = pd.concat([df, encoded_df], axis=1).drop(ohe_features, axis=1)
```

## Part 3: Calculating Proximity /Correlation Analysis of two features 

```python
def simple_matching(list1, list2):
    matches = sum(1 for a, b in zip(list1, list2) if a == b)
    return matches / len(list1)
simple_matching(df_final['Gender_M'],df_final['BikeBuyer']) 
```

```python
def jaccard(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
jaccard(df_final['Gender_M'],df_final['BikeBuyer'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
gender_m = df_final['Gender_M'].values.reshape(-1, 1)
marital_status_s = df_final['MaritalStatus_S'].values.reshape(-1, 1)

similarity = cosine_similarity(gender_m, marital_status_s)
similarity
```
